```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
df=pd.read_csv('/content/archive (1).zip')
df.head()
```

```python
df=pd.read_csv('/content/archive (1).zip',index_col=0)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.sort_index()
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Weekly_Sales'])
plt.title("Weekly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Weekly_Sales")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# sales column ka naam tumhare dataset ke hisaab se
sales_col = 'Weekly_Sales'  # ya jo bhi tumhare df me hai

# decomposition
decomposition = seasonal_decompose(df[sales_col], model='additive', period=12)

# plot
decomposition.plot()
plt.show()
```

```python
# Sales column ka naam check karna
sales_col = 'Weekly_Sales'  # ya jo bhi correct column ho

# Train-Test split
train = df[:'2017-12-31']  # 2010-2017 train
test = df['2018-01-01':]   # 2018 test

# Check
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
# Step 8: ARIMA Model Fit

from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")

# Sales column name (same as in Step 7)
sales_col = 'Weekly_Sales'  # adjust if different

# Fit ARIMA model on train data
model = ARIMA(train[sales_col], order=(1,1,1))  # train is from Step 7
model_fit = model.fit()

# Print model summary
print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=12)
forecast
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(train, label='Training Data')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.show()
```
